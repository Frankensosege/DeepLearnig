# 다중선형회귀
$$\hat{y} = a_1X_1 + a_2X_2 + \dots + b$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = np.array([2, 4, 6, 8])
X2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(X1, X2, y)

plt.show()
```

```python
a1, a2, b = (0, 0, 0)
lr = 0.01     # 학습률 설정
epoch = 2001  # 2001회 반복
n = len(X1)
```

```python
# 경사하강법

for i in range(epoch):
    y_pred = a1*X1 + a2*X2 + b
    err = y - y_pred
    a1 = a1 - lr * ((2/n) * sum(-X1 * (err)))
    a2 = a2 - lr * ((2/n) * sum(-X2 * (err)))
    b = b - lr * ((2/n) * sum(-(err)))
    
    if i%100 == 0:
        print("epoch : {},  기울기1 : {:3.2f},  기울기2 : {:3.2f}, 절편 : {}".format(i, a1, a2, b))
        
y_pred = a1 * X1 + a2 * X2 + b
print("실제 점수 : ", y)
print("예측 점수 : ", y_pred)

print("예측 점수 : ", a1*5 + a2*3 + b)
```

# 텐서플로에서 선형, 다중회귀
$$가설함수 H(x) = wx + b$$
$$최소 제곱 오차 \to 손실함수(loss function: MSE, MLE)$$
$$경사하강법 \to 옵티마이저(Optimizer: sgd, adam \dots)$$

https://www.tensorflow.org/guide/keras/sequential_model?hl=ko

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

X = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 단순 선형 회귀
model = Sequential()

model.add(Dense(1, input_dim=1, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
model.summary()
```

```python
model.fit(X, y, epochs=2000)

plt.scatter(X, y)
plt.plot(X, model.predict(X), color='r')
plt.show()
```

```python
print(model.predict([5]))
```

```python
# 다중 선형 회귀

X = np.array([[2, 0], [4, 4], [6, 2], [8,3]])
y = np.array([81, 93, 91, 97])

model1 = Sequential()

model1.add(Dense(1, input_dim=2, activation='linear'))
model1.compile(optimizer='sgd', loss='mse')
model1.summary()
model1.fit(X, y, epochs=2000)  # verbose = False : 진행상태 출력
```

```python
print(model1.predict([[7,2]]))
```

# 로지스틱회귀
$$y = \frac{1}{(1+{e}^{-(ax + b)})} : 시그모이드함수$$

$$c(H(x), y) = -log(H(x)) : y = 1 (양성 클래스)$$
$$c(H(x), y) = -log(1-H(x)) : y = 0 (음성 클래스)$$
$$-ylog(H(x))-(1-y)log(1-H(x)) : Cross Entropy 손실함수$$

```python
X = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()

model.add(Dense(1, input_dim=1, activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy')  # 이진 Cross Entropy -> 두개의 정답 class
model.fit(X, y, epochs=2000)  # verbose = False : 진행상태 출력
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), 'r')
plt.show()
```

```python
print(model.predict([7]))
```

# 다층 퍼셉트론
$$n_1=\sigma(x_1w_{11} + x_2w_{21} + b_1)$$
$$n_2=\sigma(x_1w_{12} + x_2w_{22} + b_2)$$
$$y_{out}=\sigma(n_1w_{31} + n_2x_{32} + b_3)$$

$$W^{(1)}=\begin{bmatrix}
       w_{11} & w_{12} \\[0.3em]
       w_{21} & w_{22} \\[0.3em]
     \end{bmatrix}
  B^{(1)}=\begin{bmatrix}
       b_1 \\[0.3em]
       b_2 \\[0.3em]
     \end{bmatrix}
$$
$$W^{(2)}=\begin{bmatrix}
       w_{31} \\[0.3em]
       w_{32} \\[0.3em]
     \end{bmatrix}
  B^{(2)}=\begin{bmatrix}
       b_3 \\[0.3em]
     \end{bmatrix}
$$

### XOR 문제에서 퍼셉트론을 논리게이트로 만드는 가중치
$$W^{(1)}=\begin{bmatrix}
       -2& 2 \\[0.3em]
       -2& 2 \\[0.3em]
     \end{bmatrix}
  B^{(1)}=\begin{bmatrix}
       3 \\[0.3em]
       -1\\[0.3em]
     \end{bmatrix}
$$
$$W^{(2)}=\begin{bmatrix}
       1 \\[0.3em]
       1 \\[0.3em]
     \end{bmatrix}
  B^{(2)}=\begin{bmatrix}
       -1\\[0.3em]
     \end{bmatrix}
$$


```python
# XOR 문제 해결
w11 = np.array([-2, -2])  # 처음레이어의 처음 가중치
w12 = np.array([2, 2])    # 처음레이어의 두번째 가중치
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1
```

```python
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
    
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)
    
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)
    
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
```

```python
for x in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    y = XOR(x[0], x[1])
    print('입력값 ; ', str(x), ' 출력 값 :', str(y))
```

# 오차역전파에서 딥러닝

### 오차역전파 계산법 : 교재 심화학습 1 참고
$$오차 y_{01} = \frac{1}{2}(y_{t1} - y_{01})^2 : y_{t1} \to y_{01}의 실제값$$
$$오차 y_{02} = \frac{1}{2}(y_{t2} - y_{02})^2 : y_{t2} \to y_{02}의 실제값$$
$$오차 Y_{out} = \frac{1}{2}(y_{t1} - y_{01})^2 + \frac{1}{2}(y_{t2} - y_{02})^2$$

$$\frac{\partial 오차Y_{out}}{\partial w_{31}} = \frac{\partial 오차Y_{out}}{\partial y_{01}} \cdot \frac{\partial y_{01}}{\partial 가중합_{3}} \cdot \frac{\partial 가중합_{3}}{\partial w_{31}}$$

$$\begin{align}
        (1)\quad\frac{\partial 오차Y_{out}}{\partial y_{01}} & =  \frac{\partial 오차 y_{01} + \partial 오차 y_{02}}{\partial y_{01}}\\
        & = \frac{\partial\frac{1}{2}(y_{t1} - y_{01})^2}{\partial y_{01}} + \frac{\partial\frac{1}{2}(y_{t2} - y_{02})^2}{\partial y_{01}} \\
        & = \frac{1}{2}(y_{t1} - y_{01})^2 + (y_{t2} - y_{02})^{\prime}\\
        & = (y_{t1} - y_{01})\cdot(-1)\\
        & = y_{01} - y_{t1}
\end{align}$$

$$(2)\quad\frac{\partial y_{01}}{\partial 가중합_{3}} = 활성화함수_3의 미분 : \frac{1}{(1+{e}^{-x})}$$
$$\frac{\partial \sigma(x)}{\partial x} = \sigma(x)\cdot(1-\sigma(x))$$

$$(3)\quad\frac{\partial 가중합_{3}}{\partial w_{31}}$$
$$가중합_{3} = w_{31}y_{h1} + w_{31}y_{h2} + 1(바이어스)$$
$$\therefore\quad\frac{\partial 가중합_{3}}{\partial w_{31}} = y_{h1}$$

$$w_{31}(t+1)=w_{31}t-(y_{01}-y_{t1})\cdot y_{01}(1-y_{01})\cdot y_{h1}$$
